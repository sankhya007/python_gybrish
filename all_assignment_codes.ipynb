```python
# Define the tuples
first_tuple = ("abcd", 147, 2.43, "Tom", 74.8)
small_tuple = (222, "Tom")

# a) Print complete tuple
print("Complete tuple:", first_tuple)

# b) Print the first element of the tuple
print("First element of the tuple:", first_tuple[0])

# c) Print elements starting from 2nd till 3rd
print("Elements from 2nd to 3rd:", first_tuple[1:3])

# d) Print elements starting from 3rd element
print("Elements from the 3rd element:", first_tuple[2:])

# e) Print the total length of the tuple
print("Length of the tuple:", len(first_tuple))

# f) Print complete small_tuple
print("Complete small_tuple:", small_tuple)

# g) Print tuple 2 times
print("small_tuple 2 times:", small_tuple * 2)

# h) Print concatenated two tuples
concatenated_tuple = first_tuple + small_tuple
print("Concatenated tuple:", concatenated_tuple)

# i) Program to Subtract a Tuple from Another Tuple
# Tuples in Python are immutable and subtraction operation doesn't exist for tuples directly.
# You can create a new tuple with elements that are in the first tuple but not in the second tuple.

# Example to subtract small_tuple from first_tuple
subtracted_tuple = tuple([item for item in first_tuple if item not in small_tuple])
print("Subtracting small_tuple from first_tuple:", subtracted_tuple)
```

```python
#etaa baad diyee detaa complex calculator etaa amni korechi as a experiment lagbe naah 
method = input("enter the operator (+,-,*,/,%) : ")

def addition(x,y):
    return(x+y)

def substraction(x,y):
    return(x-y)

def mulltiplication(x,y):
    return(x*y)

def divition(x,y):
    return(x/y)

def left_out(x,y):
    return(x%y)

num_1 = float(input("enter the 1st number : "))
num_2 = float(input("enter the 2nd number : "))

if method == "+":
    result = addition(num_1,num_2)
    print("the addition of the numbers {}, {} are : {}".format(num_1,num_2,result))

elif method == "-":
    result = substraction(num_1,num_2)
    print("the addition of the numbers {}, {} are : {}".format(num_1,num_2,result))

elif method == "*":
    result = mulltiplication(num_1,num_2)
    print("the addition of the numbers {}, {} are : {}".format(num_1,num_2,result))

elif method == "/":
    result = divition(num_1,num_2)
    print("the addition of the numbers {}, {} are : {}".format(num_1,num_2,result))

elif method == "%":
    result = left_out(num_1,num_2)
    print("the addition of the numbers {}, {} are : {}".format(num_1,num_2,result))
    
else:
    print("invalid input !!")
```

```python
#7)e

def find_factors(number):
    factors = []
    for i in range(1, number + 1):
        if number % i == 0:
            factors.append(i)
    return factors

number = int(input("enter a number you want to get the factorial of : "))
factors_of_number = find_factors(number)
print("Factors of", number, "are:", factors_of_number)
```

```python
#10)a

def count_strings_with_same_first_last_character(string_list):
    count = 0
    for word in string_list:
        if len(word) >= 2 and word[0] == word[-1]:
            count += 1
    return count

sample_list_strings = ['abc', 'xyz', 'aba', '1221']
print("Count of strings with same first and last character:", count_strings_with_same_first_last_character(sample_list_strings))

```

```python
#10)b

def replace_last_value_with_100(tuple_list):
    return [t[:-1] + (100,) for t in tuple_list]

sample_list_tuples = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
print(" Tuples with the last value replaced with 100:", replace_last_value_with_100(sample_list_tuples))
```
