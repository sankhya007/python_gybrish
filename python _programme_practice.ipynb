```python
import math

def cone_surface_area(radius, height):
    slant_height = math.sqrt(radius**2 + height**2)
    surface_area = math.pi * radius * (radius + slant_height)

    return surface_area

cone_radius = float(input("Enter the radius of the cone: "))
cone_height = float(input("Enter the height of the cone: "))

area = cone_surface_area(cone_radius, cone_height)
print(f"The surface area of the cone is: {area:.2f}")
```

```python
month_days = [ 0,31,28,31,30,31,30,31,31,30,31,30,31]

year = int(input("write the year : "))
month = int(input("writhe the number of the month : "))

def leap_year(year):

  return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

def days_in_month(year,month):

  if not 1 <= month <= 12 :
    return "invalid input !!"

  if month == 2 and leap_year(year):
    return 29

  return month_days[month]

print(days_in_month(year,month))
```

```python
import pickle

no_students = int(input("enter the number of students : "))
grades = []
names = []

for i in range(0,no_students,1):
    name = input("enter the name of the student : ")
    names.append(name)
    form = "enter " + name + "'s grade :"
    number = float(input(form))
    grades.append(number)
    
with open("name_grade.pkl","wb") as data_write:    
    pickle.dump(no_students,data_write)
    pickle.dump(grades,data_write)
    pickle.dump(names,data_write)

with open("name_grade.pkl","rb") as data_read:    
    a = pickle.load(data_read)
    b = pickle.load(data_read)
    c = pickle.load(data_read)
    
print("the number of students you have is : ",a)
print("here are the grades : ",b)
print("names : ",c)
```

```python
#we have to input the numbers , output them get the avg , get the high and the lowest number 

def input_num(nums):
    grades = []
    for i in range(0,nums,1):
        grade = float(input("enter the grade : "))
        grades.append(grade)
    return grades

def printGrades(nums,x):
    for i in range(0,nums,1):
        print(x[i])

def avgGrades(nums,x):
    total = 0 
    for i in range(0,nums,1):
        total = total + x[i]
    average = total/nums     
    return average
    
def topBottom(nums,x):
    topG = 0
    bottomG = 100
    for i in range(0,nums,1):
        if(x[i] < bottomG):
            bottomG = x[i]
        if(x[i] >  topG):
            topG = x[i]
    return topG,bottomG

input_grades = int(input("enter the number of grades you have : "))
myGrades = input_num(input_grades)
printGrades(input_grades,myGrades)
avg = avgGrades(input_grades,myGrades)
print("the avg of the entered functions are : ",round(avg,1))
topG,bottomG = topBottom(input_grades,myGrades)
print("your high grades is : ",topG)
print("you low grades is : ",bottomG)
```
