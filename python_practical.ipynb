```python
# Write a program in python to check whether a given number is Armstrong or not.

def is_armstrong(num):
    # Count the number of digits
    num_str = str(num)
    num_digits = len(num_str)
    
    # Calculate Armstrong sum
    armstrong_sum = sum(int(digit) ** num_digits for digit in num_str)
    
    # Check if the number is Armstrong
    return armstrong_sum == num

# Test the function
num = int(input("Enter a number: "))
if is_armstrong(num):
    print(num, "is an Armstrong number.")
else:
    print(num, "is not an Armstrong number.")
```

```python
# Write a program in python to display all the prime numbers in a given range of values.

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def prime_numbers_in_range(start, end):
    prime_numbers = []
    for num in range(start, end + 1):
        if is_prime(num):
            prime_numbers.append(num)
    return prime_numbers

# Test the function
start = int(input("Enter the starting number of the range: "))
end = int(input("Enter the ending number of the range: "))
print("Prime numbers between", start, "and", end, "are:")
print(prime_numbers_in_range(start, end))
```

```python
# Write a program in python to display the following patterns:
# A
# B C
# D E F
# G H I J
# K L M N O


def display_pattern(rows):
    current_char = ord('A')
    for i in range(1, rows + 1):
        for j in range(i):
            print(chr(current_char), end=" ")
            current_char += 1
        print()

# Test the function
rows = 5
display_pattern(rows)
```

```python
# Write the code in python to display summation of the following series:
#  s= x+ x
# 2 + x
# 3 +……….+ x
# n
# (value of x and n to be taken from user)


def series_sum(x, n):
    total_sum = 0
    for i in range(1, n + 1):
        total_sum += x ** i
    return total_sum

# Take input from the user
x = int(input("Enter the value of x: "))
n = int(input("Enter the value of n: "))

# Calculate and print the summation of the series
result = series_sum(x, n)
print("Sum of the series is:", result)
```

```python
# Write the code in python to display summation of the following series:
# s= 1 + (1+2) + (1+2+3) +……….+ (1+2+3+….+N) (value of N to be taken from user)


def series_sum(n):
    total_sum = 0
    for i in range(1, n + 1):
        inner_sum = sum(range(1, i + 1))
        total_sum += inner_sum
    return total_sum

# Take input from the user
n = int(input("Enter the value of N: "))

# Calculate and print the summation of the series
result = series_sum(n)
print("Sum of the series is:", result)
```

```python
# Write a Python code to accept a name (containing three words) and display only the initials
# (i.e. first letter of each word).
# Sample input: LAL KRISHNA ADVANI
# Sample output: L K A

def get_initials(name):
    # Split the name into words
    words = name.split()
    # Extract the first letter of each word and join them
    initials = ''.join(word[0].upper() for word in words)
    return initials

# Take input from the user
name = input("Enter a name containing three words: ")

# Get and print the initials
initials = get_initials(name)
print("Initials:", initials)

```

```python
# Write a program to input a line of text and do the following:
# i) Print the longest word from it.
#  ii)Display the count of vowels in it.
 
 
def longest_word_and_vowel_count(text):
    words = text.split()
    
    # Find the longest word
    longest_word = max(words, key=len)
    
    # Count vowels
    vowels = "aeiouAEIOU"
    vowel_count = sum(word.count(vowel) for word in words for vowel in vowels)
    
    return longest_word, vowel_count

# Take input from the user
line = input("Enter a line of text: ")

# Get the longest word and vowel count
longest_word, vowel_count = longest_word_and_vowel_count(line)

# Print the results
print("Longest word:", longest_word)
print("Vowel count:", vowel_count)
```

```python
# Write a program in Python to enter a string.
#  Check whether it is palindrome or not
 
def is_palindrome(s):
    # Convert the string to lowercase and remove non-alphanumeric characters
    s = ''.join(char.lower() for char in s if char.isalnum())
    # Check if the string is equal to its reverse
    return s == s[::-1]

# Take input from the user
string = input("Enter a string: ")

# Check if the string is a palindrome
if is_palindrome(string):
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
# Write a program in Python to enter a string.
#  Count and display number of digits, spaces and uppercase, lowercase letter.
 
def count_characters(string):
    num_digits = sum(char.isdigit() for char in string)
    num_spaces = sum(char.isspace() for char in string)
    num_uppercase = sum(char.isupper() for char in string)
    num_lowercase = sum(char.islower() for char in string)
    
    return num_digits, num_spaces, num_uppercase, num_lowercase

# Take input from the user
string = input("Enter a string: ")

# Count the characters
num_digits, num_spaces, num_uppercase, num_lowercase = count_characters(string)

# Display the counts
print("Number of digits:", num_digits)
print("Number of spaces:", num_spaces)
print("Number of uppercase letters:", num_uppercase)
print("Number of lowercase letters:", num_lowercase)
```

```python
# Write a program to take a list of integers and do the following tasks:
# a. display the largest value
# b. display the average of all the values
# c. Display the list in reverse order

def list_operations(numbers):
    # a. Display the largest value
    largest_value = max(numbers)
    print("Largest value:", largest_value)

    # b. Display the average of all the values
    average_value = sum(numbers) / len(numbers)
    print("Average value:", average_value)

    # c. Display the list in reverse order
    reversed_list = numbers[::-1]
    print("Reversed list:", reversed_list)

# Take input from the user
numbers = list(map(int, input("Enter a list of integers separated by space: ").split()))

# Perform the operations
list_operations(numbers)
```

```python
# Write a program in python to accept a list of integers and add 10 with each even element of
# the list and multiply every odd elements by 5. Display the modified list.
# Example: input: L=[10,15,12,18]
# output: L=[20,75,22,28]

def modify_list(numbers):
    modified_list = []
    for num in numbers:
        if num % 2 == 0:
            modified_list.append(num + 10)
        else:
            modified_list.append(num * 5)
    return modified_list

# Take input from the user
numbers = list(map(int, input("Enter a list of integers separated by space: ").split()))

# Modify the list
modified_list = modify_list(numbers)

# Display the modified list
print("Modified list:", modified_list)
```

```python
# Write a program to take two list L1 and L2 and perform the following tasks:
# a. Display the union of L1 and L2.
# b. Display the intersection of L1 and L2.

def list_operations(L1, L2):
    # a. Display the union of L1 and L2
    union = list(set(L1) | set(L2))
    print("Union of L1 and L2:", union)

    # b. Display the intersection of L1 and L2
    intersection = list(set(L1) & set(L2))
    print("Intersection of L1 and L2:", intersection)

# Take input from the user for two lists
L1 = list(map(int, input("Enter elements of the first list separated by space: ").split()))
L2 = list(map(int, input("Enter elements of the second list separated by space: ").split()))

# Perform the operations
list_operations(L1, L2)
```

```python
# Write a program in python to accept a list L of integer elements. Create two new list L1 and
# L2 so that L1 contains all even values and L2 contains all the odd values of L. Display L1 and
# L2.
# Example: L=[111,12,13,14,15]
# ,L1=[12,14]
# L2=[111,13,15]

def split_list(L):
    L1 = [num for num in L if num % 2 == 0]
    L2 = [num for num in L if num % 2 != 0]
    return L1, L2

# Take input from the user
L = list(map(int, input("Enter a list of integer elements separated by space: ").split()))

# Split the list into two lists containing even and odd values
L1, L2 = split_list(L)

# Display L1 and L2
print("L1 containing even values:", L1)
print("L2 containing odd values:", L2)
```

```python
# Consider a tuple of non-zero numbers. Write a program in Python to shift the negative
# numbers to right and the positive numbers to left so that the resultant list will be as follows :
# Original tuple: (-2,1,-3,-15,16,-17,5,-3,-6)
# Output should be: (1,16,5,-6,-3,-17,-15,-3,-2)

def shift_numbers(tup):
    # Separate positive and negative numbers
    positive_nums = [num for num in tup if num > 0]
    negative_nums = [num for num in tup if num < 0]
    
    # Concatenate positive numbers followed by negative numbers
    shifted_tup = tuple(positive_nums + negative_nums)
    return shifted_tup

# Original tuple
original_tuple = (-2, 1, -3, -15, 16, -17, 5, -3, -6)

# Shift the numbers and display the result
result_tuple = shift_numbers(original_tuple)
print("Output:", result_tuple)
```

```python
# Given a tuple of pairs = ((2, 5), (4, 2), (9,8), (12,10)), count the number of pairs (a, b) such
# that both a and b are even.

def count_even_pairs(pairs):
    count = 0
    for pair in pairs:
        a, b = pair
        if a % 2 == 0 and b % 2 == 0:
            count += 1
    return count

# Given tuple of pairs
pairs = ((2, 5), (4, 2), (9, 8), (12, 10))

# Count the number of pairs where both elements are even
even_pairs_count = count_even_pairs(pairs)
print("Number of pairs where both elements are even:", even_pairs_count)
```

```python
# Given a tuple of pairs = ((2, 5), (4, 2), (9,8), (12,10)), count the number of pairs (a, b) such
# that both a and b are even.

def count_even_pairs(pairs):
    count = sum(1 for a, b in pairs if a % 2 == 0 and b % 2 == 0)
    return count

# Given tuple of pairs
pairs = ((2, 5), (4, 2), (9, 8), (12, 10))

# Count the number of pairs where both elements are even
even_pairs_count = count_even_pairs(pairs)
print("Number of pairs where both elements are even:", even_pairs_count)
```

```python
# Write a program in python to take a list of some integer values. Create a dictionary from the
# list such that keys will be every elements of list and values will be cube of each element.
# Example: L=[2,3,4]
# D={2:8,3:27,4:64}

def create_cube_dictionary(lst):
    cube_dict = {num: num ** 3 for num in lst}
    return cube_dict

# Take input from the user
L = list(map(int, input("Enter a list of integer values separated by space: ").split()))

# Create the dictionary with cube values
cube_dict = create_cube_dictionary(L)

# Display the dictionary
print("Dictionary:", cube_dict)
```

```python
# Write a python program to input the name of students with their age as key-value pairs of a
# dictionary and perform the following operations:
# i) Ask user to enter any name and delete that particular pair.
# ii) Display only the name of every student.

def delete_student(dictionary, name):
    if name in dictionary:
        del dictionary[name]
        print("Pair with name '{}' deleted successfully.".format(name))
    else:
        print("Name '{}' not found in the dictionary.".format(name))

def display_names(dictionary):
    print("Names of students:")
    for name in dictionary.keys():
        print(name)

# Function to perform operations
def perform_operations(student_dict):
    # i) Delete a particular pair
    name_to_delete = input("Enter the name of the student to delete: ")
    delete_student(student_dict, name_to_delete)

    # ii) Display only the names of students
    display_names(student_dict)

# Input student names and ages
student_dict = {}
num_students = int(input("Enter the number of students: "))
for i in range(num_students):
    name = input("Enter name of student {}: ".format(i + 1))
    age = int(input("Enter age of student {}: ".format(i + 1)))
    student_dict[name] = age

# Perform operations
perform_operations(student_dict)
```

```python
# Write a Python Program to replace last value of each tuple with 100 in list. Sample List: 
#  [ (10, 20, 40) , (40, 50, 60) , (70 , 80, 90)] 
#  Expected Output: [ (10, 20, 100) , (40, 50, 100) , (70 , 80, 100)]
 
 
 def replace_last_value_with_100(lst):
    modified_list = [tuple(t[:-1] + (100,)) for t in lst]
    return modified_list

# Sample list
sample_list = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

# Replace the last value of each tuple with 100
output_list = replace_last_value_with_100(sample_list)

# Display the modified list
print("Expected Output:", output_list)
```
